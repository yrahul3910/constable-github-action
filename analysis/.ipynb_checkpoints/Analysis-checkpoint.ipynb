```python
df = pd.read_csv(io.StringIO(tbl), sep='|', header=0, skipinitialspace=True)
```

```python
df = df[df.columns[1:-1]]
```

```python
df
```

```python
df.loc[:,df.columns[0]] = [x.strip() for x in df[df.columns[0]]]
```

```python
df.loc[:,df.columns[1]] = [x.strip() for x in df[df.columns[1]]]
```

```python
df.loc[:,df.columns[0]] = [x.split('/')[-1] for x in df[df.columns[0]]]
```

```python
df
```

```python
list(df.columns)
```

```python
groups = df.groupby(by=list(df.columns)[:2], axis=0)
```

```python
for _,__ in groups:
    print(_)
    print()
    print()
    print(__)
    break
```

```python
groups['Human rating '].aggregate(np.mean)
```

```python
human_ratings = groups['Human rating '].aggregate(np.mean).unstack()
```

```python
human_ratings = human_ratings.dropna()
```

```python
plt.rcParams['figure.dpi'] = 150
human_ratings.plot(kind='bar');
```

```python
groups = df.groupby(by=list(df.columns)[0], axis=0)
```

```python
sys_ratings = groups['System rating '].aggregate(np.mean)
sys_ratings.plot(kind='bar');
```
